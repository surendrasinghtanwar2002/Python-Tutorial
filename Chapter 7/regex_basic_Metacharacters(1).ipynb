```python
                                                            ## Metacharacters are characters with a special meaning ##

                        ## This is the character class which are used to specify range of data or a group of characters you want to search in the data ##
import re
```

```python
 #[] A set of Characters  (Used to drop the special meaning of character following it)
```

```python
#Practise_no.1
common_data = "My name is Surendra Singh Tanwar and I have completed my BCA from the Poornima University. My email address is surendra@gmail.com and my contact no is 124242525252"

pattern1 = r"[abc]"                     ## It will matches  only one of the characters

result1 = re.finditer(pattern1,common_data)

print(f"This will return the regex object:- {result1}")

##Calling the for loop to iter every single value from the object

for outcome in result1:
    print(f"This is the actual data of the pattern:- {outcome}")
```

```python
#Practise 2
pattern2 = r"[^abc]"                                            ##Matches any character that is not 'a', 'b', or "c"

result2 = re.finditer(pattern2,common_data)

print(f"This is the object return through the finditer method {result2}")

##Calling the for loop and iterating all the values
for output1 in result2:
    print(f"This is the ouput {output1}")
```

```python
pattern3 = r"[0-9]"                                 ##Matches any digit from the "0" to "9"

result3 = re.finditer(pattern3,common_data)

print(f"Ojbect returned by the function {result3}")

##Calling the for loop and iterating every object value
for output3 in result3:
    print(f"This is the output of the pattern {output3}")
```

```python
pattern4 = r"[A-Z]"                             ##Matches any pattern from the Capital A to Z

result4 = re.finditer(pattern4,common_data)

print(f"This is the object returned by function {result4}")

##Calling the for loop method

for output4 in result4:
    print(f"This is the output {output4}")
```

```python
pattern5 = r"[a-z]"                             ##Matches any pattern from the small a to z

result5 = re.finditer(pattern5,common_data)

print(f"This is the object of the data {result5}")

##Calling the for loop 

for output5 in result5:
    print(f"This is the output of function {output5}")
```

```python
pattern6 = r"[A-Za-z]"                                      ## Matches any uppercase or lowercase letters ##

result6 = re.finditer(pattern6,common_data)

##Calling the data through the loop
for output6 in result6:
    print(f"This is actual data of the loop {output6}")
```

```python
pattern7 = r"[0-9A-Fa-f]"                                       ## Matches a hexadecimal digit (0-9, A-F, or a-f)

result7 = re.findall(pattern7,common_data)

for output7 in result7:
    print(f"This is the ouput of the function {output7}")
```

```python
pattern8 = r"[a-zA-Z0-9]"                                   ##Matches any alphanumeric characters

result8 = re.findall(pattern8,common_data)  

for outcome8 in result8:
    print(f"This is the output of the server {outcome8}")
```

```python
pattern9 = r"[^0-9]"                                            ##Matches any character that is not a digit

result9 = re.finditer(pattern9,common_data)     

##Calling the for loop to find the digit

for outcome9 in result9:
    print(f"This is the outcome {outcome9}")
```

```python
pattern10 = r"[^A-Za-z]"                                #This XOR ^ operator will matches any character that is not uppercase or lowercase letter

result10 = re.findall(pattern10,common_data)

##Call the loop
for outcome10 in result10:
    print(f"This is the result of the outcome {outcome10}")
```

```python
pattern11 = r"[$.]"                                         ## It will matches $ and . in the raw data ##

actual_data = "My name is $urendra $singh tanwar. I am studing regex right now so please don't disturb me"

result11 = re.findall(pattern11,actual_data)

for outcome11 in result11:
    print(f"This is the output of the file {outcome11}")
```

```python
pattern12 = r"[aieou]"                                          ## Matches vowels in the existing string ##

result12 = re.findall(pattern12,actual_data,re.IGNORECASE)

##Calling the for loop to iterate all the values
for outcome12 in result12:
    print(f"This is the volwels presented in the string {outcome12}")
```

```python
pattern13 = r"[^A-Za-z0-9]"

result13 = re.findall(pattern13,common_data)

##Calling the for loop
if result13:
    print(f"We got the result {result13}")
else:
    print(f"We have not found any result {result13}")
```

```python
                                                                        ## Practise Question ##

##Validating password using the regex and my password should contain one character or one integer value

userpass = input("Enter your password:- ")

pattern14 = r"[A-Z0-9]"

result14 = re.findall(pattern14,userpass)

if result14:
    print(f"Yes password is valid {userpass}")

else:
    print(f"Password is not valid {userpass}")
```
