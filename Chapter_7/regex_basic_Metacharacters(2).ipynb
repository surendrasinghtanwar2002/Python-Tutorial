```python
                                                            ## Metacharacters are characters with a special meaning ##

# (1) '.'	Any character (except newline character) # (2) '\' Signals a special sequence (can also be used to escape special characters)

# (3) "{}"	Exactly the specified number of occurrences

import re
```

```python
common_data = "My name is Suredra Singh Tanwar and my date of birth is 02-06-2002. My email address is surendra@gmail.com $$"
pattern1 = r"."                                     ##This will search any character in the string except new lines

result1 = re.finditer(pattern1,common_data)

##Calling the for loop
for output1 in result1:
    print(f"Starting index value:- {output1.start()} Ending index value:- {output1.end()} Grouping index value:- {output1.group()}")
```

```python
pattern2 = r"\."                                ##It will skip any special characters and only give us "." from the string

result2 = re.finditer(pattern2,common_data)

##Calling the for loop
for output2 in result2:
    print(f"Staring index value:- {output2.start()} Ending index value:- {output2.end()} Grouping value:- {output2.group()}")
```

```python
pattern3 = r"\d{2}-\d{2}-\d{4}"

result3 = re.finditer(pattern3,common_data)

##Calling the for loop
for output3 in result3:
    print(f"Starting index value {output3.start()} Ending index value {output3.end()} Grouping value {output3.group()}")
```

```python
## Opening syslog file and performing various regex operation on it

with open("E:\Python-Tutorial\Chapter_7\syslog.txt","r",encoding="utf-8") as f:
    common_Data1 = f.read()
```

```python
                                                        ## Findind date format pattern in the string ##
pattern4= r"\d{4}-\d{2}-\d{2}"              

result4 = re.finditer(pattern4,common_Data1)

##Calling the for loop to iterate every object to the memory 
for output4 in result4:
    print(f"Starting index value:- {output4.start()} Ending index value:- {output4.end()} Grouping value:- {output4.group()}")
```

```python
                                                ## Finding Time stamp pattern from the string ##
pattern5 = r"\d{2}:\d{2}:\d{2}"

result5= re.finditer(pattern5,common_Data1)

##Calling the for loop to iteratve every items presented in the object
for output5 in result5:
    print(output5.group())
```

```python
                                                            ## Finding both time stamp and date stamp from the logs ##
pattern6 = r"(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2})"

result6 = re.finditer(pattern6,common_Data1)

##Calling the for loop 
for output6 in result6:
    print(f"This is actual data {output6.group()}")
```

```python
pattern7 = r"(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2} \w{7})"

result7 = re.finditer(pattern7,common_Data1)

##Calling the for loop to iter every value
for output7 in result7:
    print(f"This is the log of the server:- {output7.group()}")
```
