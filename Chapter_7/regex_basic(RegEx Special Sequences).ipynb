```python
                    ## A special sequence is a \ followed by one of the characters in the list below, and has a special meaning ##


# Instead of using character class we can use the special character to find the #

import re
```

```python
with open("E:\\Python-Tutorial\\Chapter_7\\practise_text.txt", "r",encoding="utf-8") as file:
    # Read the contents of the file
    common_data = file.read()
```

```python
pattern1 = r"[\d]"                                      # Only matches integer or number and it will find the numerical value only

result1 = re.finditer(pattern1,common_data)

##Calling the for loop to iterate all the data from the object
for output1 in result1:
    print(f"Output index_Starting_value:-> {output1.start()} grouping:-> {output1.group()}" )
```

```python
pattern2 = r"\D"                        #Return a match where the string does not contains digit

result2 = re.finditer(pattern2,common_data)

##Calling the for loop
for output2 in result2:
    print(f"Starting index value:-> {output2.start()} and grouping:-> {output2.group()}")
```

```python
pattern3 = r"\w"                                    ## It will matches any word-character (alphanumeric plus underscore) but not include special symbol ##

result3 = re.finditer(pattern3,common_data)



##Calling the for loop
for output3 in result3:
    print(f"Starting index:-> {output3.start()} Ending index:-> {output3.end()} Grouping:-> {output3.group()}")
```

```python
pattern4 = r"\W"                ##Matches any non-word Characters

result4 = re.finditer(pattern4,common_data)

##Calling the for loop 
for output4 in result4:
    print(f"Starting index value {output4.start()} Ending index value {output4.end()} Grouping value {output4.group()} ")
```

```python
pattern5 = r"\s"                ##Matches any whitespaces characters (spaces,tabs,newlines,etc.)

result5 = re.finditer(pattern5,common_data)

##Calling the for loop
for output5 in result5:
    print(f"Staring index value {output5.start()} Ending index value {output5.end()} Grouping........-> {output5.group()}")
```

```python
pattern6 = r"\S"               ##Matches any non whitespaces characters

result6 = re.finditer(pattern6,common_data)

##Calling the for loop
for output6 in result6:
    print(f"Starting index value {output6.start()} Ending index value {output6.end()} Grouping {output6.group()}")
```
